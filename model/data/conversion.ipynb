## 5 BSs

### Aggregate nodes by number of links from the core

```python
import json

with open('./T2_5_links.json', 'r') as node_file:
    json_input = node_file.read()

link_data = json.loads(json_input)
node_levels = {0: [0], 1: []}
current_node_level = 1
upper_node_level = node_levels[current_node_level-1].copy()

for link in link_data['links']:

    # if link['fromNode'] in node_levels[current_node_level]:
    #     continue

    if link['fromNode'] not in upper_node_level:
        # -- We changed level --
        current_node_level += 1
        upper_node_level += node_levels[current_node_level-1]
        node_levels[current_node_level] = []

    # -- We are in current level --
    # Adding destination node to current level list
    if link['toNode'] not in node_levels[current_node_level]:
        node_levels[current_node_level].append(link['toNode'])

node_levels
```
